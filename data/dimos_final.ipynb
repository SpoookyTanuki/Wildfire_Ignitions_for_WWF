```python
# Read the Excel file with Fire Investigation Causes
fire_investigation_df = pd.read_excel(r'C:\Users\elpa22\Desktop\4_WWF\GIS\municipalities\Fire_Investigation_Causes_2000-2023.xlsx')

fire_investigation_df.head()
```

```python
fire_investigation_df
```

```python
electricity_causes = [
    "Στατικός ηλεκτρισμός",
    "Βραχυκύκλωμα ηλεκτρικών καλωδίων",
    "ΒΡΑΧΥΚΛΩΜΑ ΣΤΟ ΔΙΚΤΥΟ ΤΗΣ ΑΠΌ ΠΤΗΝΟ",
    "ΒΡΑΧΥΚΥΚΛΩΜΑ ΣΕ ΥΠΟΣΤΑΘΜΟ ΤΗΣ ΔΕΗ",
    "ΠΤΩΣΗ ΔΕΝΔΡΟΥ ΣΕ ΗΛΕΚΤΡΟΦΟΡΟ ΚΑΛΩΔΙΟ",
    "ΠΤΩΣΗ ΗΛΕΚΤΡΙΚΩΝ ΚΑΛΩΔΙΩΝ",
    "Υπερθέρμανση-πυράκτωση-τήξη ηλεκτρικών αγωγών",
    "Πτώση καλωδίου ΔΕΗ",
    "ΒΡΑΧΥΚΛΩΜΑ ΚΑΛΩΔΙΩΝ ΔΕΗ",
    "ΠΤΩΣΗ ΗΛΕΚΤΡΙΚΟΥ ΑΓΩΓΟΥ Δ.Ε.Η."
]
```

```python
electricity_fires_df = fire_investigation_df[
    (fire_investigation_df['ΓΕΝΙΚΗ ΑΙΤΙΑ ΠΥΡΚΑΓΙΑΣ'] == 'ΗΛΕΚΤΡΙΣΜΟΣ')
    | (fire_investigation_df['ΑΝΑΛΥΣΗ ΑΙΤΙΑΣ'].isin(electricity_causes))
]

electricity_fires_df.head()
```

```python
electricity_fires_df
```

```python
# Show count of filtered rows
print(f"\nNumber of fires caused by electricity: {len(electricity_fires_df)}")
```

```python
# Merge the English names into electricity_fires_df
electricity_fires_df = electricity_fires_df.merge(
    municipality_count_df[['ΔΗΜΟΣ', 'Name in English']], 
    on='ΔΗΜΟΣ', 
    how='left'
)
```

```python
electricity_fires_df.head()
```

```python
# Optional: Check if any municipalities didn't get matched
unmatched = electricity_fires_df[electricity_fires_df['Name in English'].isna()]
if len(unmatched) > 0:
    print("\nUnmatched municipalities:")
    print(unmatched['ΔΗΜΟΣ'].unique())
```

```python
electricity_fires_df
```

```python
electricity_fires_df = electricity_fires_df.merge(
    municipality_count_df[['ΔΗΜΟΣ', 'code']], 
    on='ΔΗΜΟΣ', 
    how='left'
)
```

```python
electricity_fires_df
```

```python
# Count unique values in each column
name_count = electricity_fires_df['Name in English'].nunique()
code_count = electricity_fires_df['code'].nunique()

print(f"Number of unique English names: {name_count}")
print(f"Number of unique codes: {code_count}")
```

```python
# Create a table with counts per code
code_counts = electricity_fires_df['code'].value_counts().reset_index()
code_counts.columns = ['Code', 'Number of Fires']
code_counts = code_counts.sort_values('Code')
```

```python
code_counts
```

```python
# Import geopandas
import geopandas as gpd
```

```python
# Read the shapefile
gdf = gpd.read_file(r'C:\Users\elpa22\Desktop\4_WWF\GIS\municipalities\oria_wgs\oria.shp')
```

```python
# Merge the fire counts with the shapefile based on the 'code' column
gdf = gdf.merge(code_counts, how='left', left_on='kallikra_1', right_on='Code')
```

```python
# Fill any NaN values with 0 (for municipalities with no fires)
gdf['Number of Fires'] = gdf['Number of Fires'].fillna(0)
```

```python
# Save the updated shapefile
gdf.to_file(r'C:\Users\elpa22\Desktop\4_WWF\GIS\municipalities\oria_el_ignitions\oria_el_ignitions.shp')
```
