```python
import pandas as pd
```

```python
file_path = r"C:\Users\elpa22\Desktop\4_WWF\from_elias\Fire_Investigation_Causes_ 2000-2023_.xlsx"
```

```python
df = pd.read_excel(file_path)
```

```python
print(df.head())
```

```python
# Count unique names in the column 'ΔΗΜΟΣ'
unique_names = df['ΔΗΜΟΣ'].nunique()
print(f"Number of unique names in ΔΗΜΟΣ: {unique_names}")
```

```python
unique_names_list = df['ΔΗΜΟΣ'].unique()
print(unique_names_list)
```

```python
# Show frequency of each municipality name
df['ΔΗΜΟΣ'].value_counts().sort_index()
```

```python
# Convert NaN values to strings and then sort
clean_municipalities = df['ΔΗΜΟΣ'].fillna('Not Specified').unique()
pd.Series(sorted(clean_municipalities)).to_frame('Municipality Names')
```

```python
# Show municipality names with their frequency counts
municipality_counts = df['ΔΗΜΟΣ'].fillna('Not Specified').value_counts().sort_index()
municipality_counts.to_frame('Count').reset_index().rename(columns={'index': 'Municipality Names'})
```

```python
# Drop NaN values before sorting
clean_municipalities = df['ΔΗΜΟΣ'].dropna().unique()
pd.Series(sorted(clean_municipalities)).to_frame('Municipality Names')
```

```python
# Show municipality names with their frequency counts
municipality_counts = df['ΔΗΜΟΣ'].fillna('Not Specified').value_counts().sort_index()
municipality_counts.to_frame('Count').reset_index().rename(columns={'index': 'Municipality Names'})
```

```python
# Set pandas to show all rows
pd.set_option('display.max_rows', None)

# Show all municipality names with their frequency counts
municipality_counts = df['ΔΗΜΟΣ'].fillna('Not Specified').value_counts().sort_index()
result = municipality_counts.to_frame('Count').reset_index().rename(columns={'index': 'Municipality Names'})
display(result)
```

```python
# Export to Excel
result.to_excel(r'C:\Users\elpa22\Desktop\4_WWF\municipality_counts.xlsx', index=False)
```
